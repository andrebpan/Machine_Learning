Arrays de N-Dimensões

```python
import numpy as np

# Lista de listas representando os elementos da matriz
lst = [[1, 2, 3], [4, 5, 6]]

# Criando a matriz bidimensional usando NumPy
ary2d = np.array(lst)

# Imprimindo a matriz resultante
print(ary2d)
```

```python
# Convertendo o tipo de dados do ndarray para float32
float32_ary = ary2d.astype(np.float32)

# Imprimindo o ndarray convertido e verificando seu tipo de dados
print(float32_ary)
print(float32_ary.dtype)
```

```python
# Determinando o tamanho dos elementos em bytes
element_size = ary2d.itemsize
print("Tamanho dos elementos em bytes:", element_size)

# Determinando o número total de elementos no array
total_elements = ary2d.size
print("Número total de elementos:", total_elements)

# Determinando o número de dimensões do array
num_dimensions = ary2d.ndim
print("Número de dimensões:", num_dimensions)

# Determinando a forma do array (número de elementos ao longo de cada dimensão)
array_shape = ary2d.shape
print("Forma do array:", array_shape)
print(np.array([1,2,3]).shape)
```

Matriz matemática e funções universais

```python
# Lista de listas representando os elementos da matriz
lst = [[1, 2, 3], [4, 5, 6]]
# Iterando sobre cada linha da matriz
for row_idx, row_val in enumerate(lst):
    # Iterando sobre cada elemento da linha
    for col_idx, col_val in enumerate(row_val):
        # Incrementando o valor de cada elemento da matriz em 1
        lst[row_idx][col_idx] += 1

# Imprimindo a matriz após a adição de 1 a cada elemento
lst
```

```python
# Usando list comprehension para adicionar 1 a cada elemento da matriz
[[cell + 1 for cell in row] for row in lst]
```

```python
# Criando um array NumPy
ary = np.array([[1, 2, 3], [4, 5, 6]])

# Adicionando 1 a cada elemento do array usando a ufunc np.add()
ary = np.add(ary, 1)

# Imprimindo o array resultante
print(ary)
```

```python
# Adicionando 1 a cada elemento do array usando o operador '+'
ary_resultante = ary + 1

# Imprimindo o array resultante
print(ary_resultante)
```

```python
# Elevando cada elemento do array ao quadrado usando o operador '**'
ary_resultante = ary ** 2

# Imprimindo o array resultante
print(ary_resultante)
```

```python
# Criando um array NumPy
ary = np.array([[1, 2, 3], [4, 5, 6]])

# Calculando as somas das colunas usando np.add.reduce()
np.add.reduce(ary)
```

```python
# Calculando as somas das linhas usando np.add.reduce() com axis=1
np.add.reduce(ary, axis=1)
```

```python
# Calculando as somas das colunas usando ary.sum(axis=0)
ary.sum(axis=0)
```

```python
# Calculando a soma de todos os elementos do array usando ary.sum()
ary.sum()
```

Operadores de comparação e máscaras

```python
# Criando uma array unidimensional
ary = np.array([1, 2, 3, 4])

# Criando uma máscara booleana onde os elementos são maiores que 2
mask = ary > 2

# Exibindo a máscara booleana
mask
```

```python
# Usando a máscara para selecionar elementos da array
ary[mask]
```

```python
mask
```

```python
# Calculando a soma dos valores True na máscara
mask.sum()
```

```python
# Usando np.where() para criar uma nova array baseada em uma condição
np.where(ary > 2, 1, 0)
```

```python
# Criando uma máscara booleana onde os elementos são maiores que 2
mask = ary > 2

# Atribuindo o valor 1 para os elementos onde a máscara é True
ary[mask] = 1

# Atribuindo o valor 0 para os elementos onde a máscara é False
ary[~mask] = 0

# Exibindo a array modificada
ary
```

```python
# Criando uma array unidimensional
ary = np.array([1, 2, 3, 4])

# Criando uma máscara booleana usando operadores lógicos
mask = (ary > 3) | (ary < 2)

# Exibindo a máscara booleana
mask
```

```python
# Criando uma máscara booleana usando operadores lógicos e negando-a
mask = ~((ary > 3) | (ary < 2))

# Exibindo a máscara booleana negada
mask
```

Álgebra Linear com Matrizes NumPy

```python
# Criando uma array unidimensional (vetor)
row_vector = np.array([1, 2, 3])

# Exibindo a array unidimensional (vetor)
row_vector
```

```python
# Criando uma array unidimensional (vetor) e, em seguida, redimensionando-a para uma coluna usando reshape
column_vector = np.array([[1, 2, 3]]).reshape(-1, 1)

# Exibindo a array unidimensional (vetor) que representa uma coluna
column_vector
```

```python
# Vetor unidimensional (vetor) original
row_vector = np.array([1, 2, 3])

# Convertendo o vetor unidimensional em um vetor de coluna usando np.newaxis
column_vector = row_vector[:, np.newaxis]

# Exibindo o vetor de coluna resultante
column_vector
```

```python
# Matriz bidimensional
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])

# Vetor de coluna
column_vector = np.array([[1], [2], [3]])

# Multiplicação de matriz entre a matriz e o vetor de coluna
result = np.matmul(matrix, column_vector)

# Exibindo o resultado da multiplicação de matriz
result
```

```python
# Multiplicação de matriz entre a matriz e o vetor coluna
np.matmul(matrix, column_vector)
```

```python
# Calculando o produto escalar de dois vetores
np.dot(row_vector, row_vector)
```

```python
# Calculando o produto escalar entre a matriz e o vetor
np.dot(matrix, row_vector)
```

```python
# Calculando o produto escalar entre a matriz e o vetor coluna
np.dot(matrix, column_vector)
```

```python
# Matriz bidimensional
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])

# Transpondo a matriz
transposed_matrix = matrix.transpose()

# Exibindo a matriz transposta
transposed_matrix
```

```python
# Calculando o produto matricial entre a matriz e sua transposta
np.matmul(matrix, matrix.transpose())
```

Histograms

```python
# Criando um gerador de números aleatórios com semente fixa
rng = np.random.RandomState(123)

# Gerando uma amostra de dados normalmente distribuídos
x = rng.normal(0, 20, 1000)

# Definindo os intervalos (bins) para o histograma
bins = np.arange(-100, 100, 5)

# Plotando o histograma
plt.hist(x, bins=bins, edgecolor='black')
# Exibindo o histograma
plt.show()
```

```python
# Criando um gerador de números aleatórios com semente fixa
rng = np.random.RandomState(123)

# Gerando duas amostras de dados normalmente distribuídos
x1 = rng.normal(0, 20, 1000)
x2 = rng.normal(15, 10, 1000)

# Definindo os intervalos (bins) para os histogramas
bins = np.arange(-100, 100, 5)

# Plotando os histogramas para as duas amostras
plt.hist(x1, bins=bins, alpha=0.5, label='x1')
plt.hist(x2, bins=bins, alpha=0.5, label='x2')

# Exibindo o gráfico
plt.show()
```
